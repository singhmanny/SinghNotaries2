```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the feedback data from CSV file
data = pd.read_csv('Feedback Data.csv')
```

```python
# Table 1: Top 5 Service Types
top_service_types = data['service_type'].value_counts().head(5)
print("Table 1: Top 5 Service Types")
print(top_service_types)
```

```python
# Table 2: Average Satisfaction by Gender
avg_satisfaction_by_gender = data.groupby('gender')['satisfaction'].mean()
print("\nTable 2: Average Satisfaction by Gender")
print(avg_satisfaction_by_gender)
```

```python
# Table 3: Likert Scale Distribution
likert_scale_distribution = data['likert_scale'].value_counts().sort_index()
print("\nTable 3: Likert Scale Distribution")
print(likert_scale_distribution)
```

```python
# Table 4: Service Rating Distribution
service_rating_distribution = data['service_rating'].value_counts().sort_index()
print("\nTable 4: Service Rating Distribution")
print(service_rating_distribution)
```

```python
# Table 5: Average Age by Service Type
avg_age_by_service_type = data.groupby('service_type')['age'].mean()
print("\nTable 5: Average Age by Service Type")
print(avg_age_by_service_type)
```

```python
# Graph 1: Bar Chart of Service Types
plt.figure(figsize=(8, 6))
data['service_type'].value_counts().plot(kind='bar')
plt.title('Service Type Distribution')
plt.xlabel('Service Type')
plt.ylabel('Count')
plt.show()
```

```python
# Graph 2: Pie Chart of Gender Distribution
plt.figure(figsize=(8, 6))
data['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Graph 3: Histogram of Age Distribution
plt.figure(figsize=(8, 6))
data['age'].plot(kind='hist', bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Graph 4: Bar Chart of Satisfaction by Gender
plt.figure(figsize=(8, 6))
data.groupby('gender')['satisfaction'].mean().plot(kind='bar')
plt.title('Average Satisfaction by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Satisfaction')
plt.show()
```

```python
# Line Chart: Age vs. Satisfaction
plt.figure(figsize=(8, 6))
plt.plot(data['age'], data['satisfaction'], marker='o', linestyle='-')
plt.title('Age vs. Satisfaction')
plt.xlabel('Age')
plt.ylabel('Satisfaction')
plt.grid(True)
plt.show()
```
